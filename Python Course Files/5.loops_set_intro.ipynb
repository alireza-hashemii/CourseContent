# For Loop

1.  Iterate ove numbers from 1 to 3

```python
for number in [1, 2, 3]:
    print(number)
```

2 . Sum all numbers in a list

```python
numbers = [2, 4, 6, 8]
total = 0

for num in numbers:
    total += num

print(total)
```

3 Print even numbers from 0 to 20

```python
for i in range(0, 21):
    if i % 2 == 0:
        print(i)
```

4. Print a pattern of stars

```python
for i in range(1, 6):
    print("*" * i)
```

5. Create a new list with squared values

```python
numbers = [1, 2, 3, 4]
squares = []

for num in numbers:
    squares.append(num ** 2)

print(squares)
```

6. Reverse a list without using .reverse() or slicing

```python
numbers = [1, 2, 3, 4, 5]
reversed_list = []

for i in range(len(numbers) - 1, -1, -1):
    reversed_list.append(numbers[i])

print(reversed_list)
```

7. Find the treasure game

```python
matrix = [
    [None, None, True],
    [None, None, None],
    [None, None, None]
]

for i in range(3):
    row = int(input("row number: "))
    column = int(input("column number: "))

    if matrix[row - 1][column - 1] == True:
        print("Found the treasure")
        break
```

# Raw String: 
**A raw string in Python is a string where escape characters are NOT processed.
Python treats backslashes (\) as literal characters, not special ones.**

```python
print("Hello\nWorld")
```

```python
print(r"Hello\nWorld")
```

# Set 

| Method                                   | Description                                           |
| ---------------------------------------- | ----------------------------------------------------- |
| `add(elem)`                              | Adds an element to the set                            |
| `clear()`                                | Removes all elements from the set                     |
| `copy()`                                 | Returns a shallow copy of the set                     |
| `difference(other_set)`                  | Returns elements in the set but not in another set    |
| `difference_update(other_set)`           | Removes elements found in another set                 |
| `discard(elem)`                          | Removes an element if it exists (no error if missing) |
| `intersection(other_set)`                | Returns common elements between sets                  |
| `intersection_update(other_set)`         | Keeps only common elements                            |
| `isdisjoint(other_set)`                  | Returns `True` if sets have no elements in common     |
| `issubset(other_set)`                    | Returns `True` if set is a subset of another          |
| `issuperset(other_set)`                  | Returns `True` if set is a superset of another        |
| `pop()`                                  | Removes and returns a random element                  |
| `remove(elem)`                           | Removes an element (raises error if missing)          |
| `symmetric_difference(other_set)`        | Returns elements in either set, but not both          |
| `symmetric_difference_update(other_set)` | Updates set with symmetric difference                 |
| `union(other_set)`                       | Returns all unique elements from both sets            |
| `update(other_set)`                      | Adds elements from another set                        |


1. Check if a password has unique characters

```python
password = "secure123"

if len(password) == len(set(password)):
    print("All characters are unique")
else:
    print("Duplicate characters found")
```

2. Remove banned words from a sentence

```python
sentence = "python is fun and python is powerful"
banned = {"idiot", "monkey"}

words = set(sentence.split())
cleaned = words - banned

print(cleaned)
```
